```python
def create_profile(name, age, email):
  if not (check_name(name) and check_email(email) and check_age(age)):
    print("You have failed the checks")
  else:
    print("Wait while we create a profile...")



def check_name(name):
  for char in name:
    if char in "!@£$%^&*()":
      print("Name cannot contain '@£$%^&*()' ")
      return False
  return True
def check_email(email):
  if "@" in email:
    return True
  else:
    print("Please enter a valid email address")
    return False

def check_age(age):
  if age > 12:
    return True
  else:
    print("You must be older than 12 years old")
    return False

create_profile("sam@", 13, "sam@gmx.com")
```

Factorial function

```python
def factorial(n):
  if n == 0:
    return 1
  else:
    return n * factorial(n-1)

factorial(5)
```

Recursive Fibonacci function

```python
def fibonacci_sequence(n):
  f_sequence = []
  for num in range(1, n+1):
    if num == 1:
      f_sequence.append(0)
    elif num == 2:
      f_sequence.append(1)
    else:
 #     print(num, f_sequence)
      f_num = sum(fibonacci_sequence(num-1)[-2:])
  #     print(num, f_num)
      f_sequence.append(f_num)
  return f_sequence

fibonacci_sequence(30)


```

```python
import string
def palindrome_no_case(my_string):
    my_string = my_string.lower().translate(str.maketrans('', '', string.punctuation))
    if my_string == my_string[::-1]:
        return True
    else:
        return False
```
